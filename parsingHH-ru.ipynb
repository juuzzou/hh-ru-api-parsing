```python
import requests
import json
from tqdm import tqdm
import pandas as pd
from collections import Counter
```

```python
vacancies = []

for i in range(0, 2):
    vac = requests.get(f"https://api.hh.ru/vacancies?text=data%20analyst&only_with_salary=true&per_page=100&page={i}")
    vacancies.extend(json.loads(vac.text)['items'])
```

```python
vacancies[0]
```

```python
for v in vacancies:
    print(v['name'], v['employer']['name'], v['area']['name'], v['url'])
```

```python
links = []

for v in vacancies:
    links.append(v['url'])

links
```

```python
info = []

for l in tqdm(links):
    v = requests.get(l)
    v_loaded = json.loads(v.text)
    
    try:
        if v_loaded['salary']['currency'] == 'RUR':
            description = v_loaded['description']
            skills = [i['name'] for i in v_loaded['key_skills']]
            schedule = v_loaded['schedule']['name']
            url = v_loaded['alternate_url']
            
            info.append([description, skills, schedule, url])
    except Exception as e:
        continue
```

```python
info
```

```python
df = pd.DataFrame(info, columns=['description', 'skills', 'schedule', 'url'])
```
